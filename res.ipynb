```python
import matplotlib.pyplot as plt
```

```python
res_multi = [3, 2,4,3]
res_milti2 = [2,1,4,2]
res_pbl = [3,0,3,2]
```

```python
TASK_NAMES = ['Align', 'Stack', 'Color Match', 'Color MisMatch','Average']
NUMS = 5
```

```python
total_score_multi1 = sum(res_multi)/(NUMS*len(res_multi))
total_score_multi2 = sum(res_milti2)/(NUMS*len(res_milti2))
total_score_pbl = sum(res_pbl)/(NUMS*len(res_pbl))
```

```python
# plot bar chart
plt.figure(figsize=(8, 6))
plt.title('Task Success Rate')
plt.ylabel('Success Rate')
plt.xlabel('Task')
plt.ylim(0, 1)
plt.bar(range(5), [i/NUMS for i in res_pbl]+[total_score_pbl], label='PBL', color='lightcoral',width=0.3)
plt.bar([i+0.3 for i in range(5)], [i/NUMS for i in res_milti2]+[total_score_multi2], label='MultiModal', 
        color = 'lightgreen', tick_label=TASK_NAMES, width=0.3)
plt.bar([i+0.6 for i in range(5)], [i/NUMS for i in res_multi]+[total_score_multi1], label='MultiModal-Ret', color='lightskyblue',
          width=0.3)
plt.plot([3.8,3.8], [0,1], color='black', linestyle='--')
plt.legend()
plt.show()
```
